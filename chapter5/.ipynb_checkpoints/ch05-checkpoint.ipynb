5-1 银行贷款拖欠率逻辑回归

```python
import pandas as pd

filename = path + '/data/bankloan.xls'
data = pd.read_excel(filename)
data.head()
```

```python
data.info()    # 无缺失值
```

```python
data.describe()
```

```python
# 划分自变量与因变量：前8列为自变量，最后一列为因变量
x = data.iloc[:, :8].as_matrix()
y = data.iloc[:, 8].as_matrix()

# 导入线性回归包：逻辑斯蒂回归（LR、RLR）
from sklearn.linear_model import LogisticRegression as LR
from sklearn.linear_model import RandomizedLogisticRegression as RLR

# 建模：随机逻辑斯蒂回归模型
rlr = RLR()
rlr.fit(x, y)
rlr_support = rlr.get_support()    # 获取特征筛选结果
support_col = data.drop('违约', axis = 1).columns[rlr_support]

print(u'通过随机逻辑斯蒂回归模型选择特征结束！')
print(u'有效特征为：%s' % ','.join(support_col))
x = data[support_col].as_matrix()    # 筛选好的特征

# 建模：逻辑斯蒂回归模型
lr = LR()
lr.fit(x, y)
print(u'逻辑回归模型训练结束！')
print(u'模型平均正确率：%s' % lr.score(x, y))
```

5-3 神经网络算法预测销量高低

```python
inputfile = path + '/data/sales_data.xls'
data = pd.read_excel(inputfile, index_col = u'序号')

# 将数据类别标签转化为数值型{0, +1}
data[data == u'是'] = 1
data[data == u'高'] = 1
data[data == u'好'] = 1
data[data != 1] = 0

# 划分自变量与因变量
x = data.iloc[:, :3].as_matrix().astype(int)
y = data.iloc[:, -1].as_matrix().astype(int)

# 导入深度学习模块
from keras.layers.core import Activation, Dense
from keras.models import Sequential
import matplotlib.pyplot as plt
%matplotlib inline

# 进行神经网络训练
model = Sequential()
model.add(Dense(input_dim = 3, units = 10))    # 添加隐层
model.add(Activation('relu'))                  # 提升准确度
model.add(Dense(input_dim = 10, units = 1))
model.add(Activation('sigmoid'))               # 0-1输出
# 修正神经网络：二元分类
model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
model.fit(x, y, batch_size = 10, epochs = 1000)    # 训练1000次

yp = model.predict_classes(x).reshape(len(y))    # 分类预测

# 绘制混淆矩阵
from sklearn.metrics import confusion_matrix
def cm_plot(y, yp):
    cm = confusion_matrix(y, yp)

    plt.matshow(cm, cmap = plt.cm.Greens)
    plt.colorbar()

    for x in range(len(cm)):
        for y in range(len(cm)):
            plt.annotate(
                cm[x, y],
                xy = (x, y),
                horizontalalignment = 'center',
                verticalalignment = 'center')

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    return plt

cm_plot(y, yp).show()
```

5-6 Apriori算法获得超市销售关联规则

```python
inputfile = path + '/data/apriori.txt'
outputfile = path + '/tmp/apriori_rules.xls'
data = pd.read_csv(inputfile, header = None, dtype = object)
data.head()
```

```python
data.info()
```

```python
# 自定义连接函数
# 实现k-1阶频繁项集L_{k-1}与1阶频繁项集L_{1}得到k阶候选集C_k的连接
# 返回k阶频繁项集L_{k}
def connect_string(x, ms):
    '''
    输入：支持度大于最小支持度的数据索引列表，连接符
    输出：下一阶频繁项集
    '''
    x = list(map(lambda i : sorted(i.split(ms)), x))    # 排序
    l = len(x[0])
    r = []
    for i in range(len(x)):
        for j in range(i, len(x)):
            if x[i][: l - 1] == x[j][: l - 1] and x[i][l - 1] != x[j][l - 1]:
                r.append(x[i][: l - 1] + sorted([x[j][l - 1], x[i][l - 1]]))
    return r

# 寻找关联规则的函数
def find_rule(data, support, confidence, ms = u'----'):
    # 定义输出结果
    result = pd.DataFrame(index = ['support', 'confidence'])
    # 支持度序列
    support_series = 1.0 * data.sum() / len(data)
    # 支持度第一次筛选，筛选出大于最小支持度的数据
    column = list(support_series[support_series > support].index)  # 索引
    k = 0

    while len(column) > 1:
        k = k + 1
        print(u'\n正在进行第%s次搜索...' % k)
        column = connect_string(column, ms)
        print(u'数目：%s...' % len(column))

        # 新一批支持度的计算函数
        sf = lambda i: data[i].prod(axis = 1, numeric_only = True)    #新一批支持度的计算函数

        # 创建连接数据，这一步耗时、耗内存最严重。当数据集较大时，可以考虑并行运算优化。
        d_2 = pd.DataFrame(list(map(sf, column)), index = [ms.join(i) for i in column]).T

        # 计算连接后的支持度
        support_series_2 = 1.0 * d_2[[ms.join(i) for i in column]].sum() / len(d)
        column = list(support_series_2[support_series_2 > support].index)  # 新一轮支持度筛选
        support_series = support_series.append(support_series_2)
        column2 = []

        for i in column:  # 遍历可能的推理，如{A,B,C}究竟是A+B-->C还是B+C-->A还是C+A-->B？
            i = i.split(ms)
            for j in range(len(i)):
                column2.append(i[:j] + i[j + 1:] + i[j:j + 1])

        cofidence_series = pd.Series(index = [ms.join(i) for i in column2])  # 定义置信度序列
        # 计算置信度序列
        for i in column2:
            cofidence_series[ms.join(i)] = support_series[
                ms.join(sorted(i))] / support_series[ms.join(i[:len(i) - 1])]
        # 置信度筛选
        for i in cofidence_series[cofidence_series > confidence].index:
            result[i] = 0.0
            result[i]['confidence'] = cofidence_series[i]
            result[i]['support'] = support_series[ms.join(sorted(i.split(ms)))]

    result = result.T.sort_values(
        ['confidence', 'support'], ascending = False)  # 结果整理，输出
    print(u'\n结果为：')
    print(result)

    return result
```

```python
import time
start = time.clock()

print(u'\n转换原始数据至0-1矩阵...')
ct = lambda x : pd.Series(1, index = x)
b = map(ct, data.as_matrix())             # map执行
data = pd.DataFrame(list(b)).fillna(0)    # 实现矩阵转换，用0填充空值

end = time.clock()
print(u'\n转换完毕! \n用时：%0.2f秒' % (end - start))
print(u'\n开始搜索关联规则...')
del b

support = 0.06       # 最小支持度
confidence = 0.75    # 最小置信度
ms = '----'          # 连接符

start = time.clock()
find_rule(data, support, confidence, ms).to_excel(outputfile)

end = time.clock()
print(u'\n搜索关联规则完毕! \n用时：%0.2f秒' % (end - start))
```

```python
# 时序图
plt.rcParams['font.sans-serif'] = ['SimHei']    # 正常显示黑体中文
plt.rcParams['axes.unicode_minus'] = False      # 正常显示负号
data.plot()
plt.title(u'差分前的时序图')
plt.show()

# 自相关图
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(data).show()

# 平稳性检测
from statsmodels.tsa.stattools import adfuller as ADF
print(u'原始序列的ADF检验结果为：', ADF(data[u'销量']))
# 返回值依次为adf、pvalue、usedlag、nobs、critical values、icbest、regresults、resstore

# 差分后的结果
D_data = data.diff(periods = 1, axis = 0).dropna()
D_data.columns = [u'销量差分']
# 时序图
D_data.plot()
plt.title(u'一阶差分后的时序图')
plt.show()
    
# 转化为差分平稳序列后的自相关图
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(D_data).show()
# 转化为差分平稳序列后的偏自相关图
from statsmodels.graphics.tsaplots import plot_pacf
plot_pacf(D_data).show()
# 差分后的平稳性检测
print(u'差分序列的ADF检验结果为：', ADF(D_data[u'销量差分']))

# 白噪声检验
from statsmodels.stats.diagnostic import acorr_ljungbox
print(u'差分序列的白噪声检验结果为：', acorr_ljungbox(D_data, lags = 1))    # 返回统计量和p值
data[u'销量'] = data[u'销量'].astype('float64')
```

```python
# 定阶
from statsmodels.tsa.arima_model import ARIMA
pmax = int(len(D_data) / 10)  # 一般阶数不超过length/10
qmax = int(len(D_data) / 10)  # 一般阶数不超过length/10
bic_matrix = []  # bic矩阵
data.dropna(inplace = True)

# 存在部分报错，所以用try来跳过报错；
# 存在warning，暂未解决使用warnings跳过
import warnings
warnings.filterwarnings('error')
for p in range(pmax + 1):
    tmp = []
    for q in range(qmax + 1):
        try:
            tmp.append(ARIMA(data, (p, 1, q)).fit().bic)
        except:
            tmp.append(None)
        bic_matrix.append(tmp)

# 从中可以找出最小值
bic_matrix = pd.DataFrame(bic_matrix)
# 用stack展平，然后用idxmin找出最小值位置。
p, q = bic_matrix.stack().idxmin()
print(u'BIC最小的p值和q值为：%s, %s' % (str(p), str(q)))
model = ARIMA(data, (p, 1, q)).fit()  # 建立ARIMA(p, d, q)模型，其中d为差分阶数
print(u'模型报告：\n', model.summary2())  # 给出一份模型报告
print(u'预测结果：\n', model.forecast(forecastnum))  # 作为期5天的预测，返回预测结果、标准误差、置信区间。
```

5-8 基于聚类的离散点检测

```python
inputfile = path + '/data/consumption_data.xls'
data = pd.read_excel(inputfile, index_col = 'Id')
data_zs = 1.0 * (data - data.mean()) / data.std()    # 数据标准化
data_zs.info()
```

```python
data_zs.head()
```

```python
# 导入KMeans聚类算法
import numpy as np
from sklearn.cluster import KMeans
k = 3               # 聚类中心数目
threshold = 2       # 离散点阈值
iteration = 1000    # 聚类最大循环次数
model = KMeans(n_clusters = k, n_jobs = 4, max_iter = iteration)
model.fit(data_zs)

# 标准化数据及其类别
r = pd.concat([data_zs, pd.Series(model.labels_, index = data.index)], axis = 1)
# 每个样本对应的类别
r.columns = list(data.columns) + [u'聚类类别']

norm = []
for i in range(k):    # 逐一处理
    norm_tmp = r[['R', 'F', 'M']][r[u'聚类类别'] == i] - model.cluster_centers_[i]
    # 求出绝对距离
    norm_tmp = norm_tmp.apply(np.linalg.norm, axis = 1)
    # 求相对距离并添加
    norm.append(norm_tmp / norm_tmp.median())
norm = pd.concat(norm)    # 合并

plt.rcParams['font.sans-serif'] = ['SimHei']  # 正常显示黑体中文
plt.rcParams['axes.unicode_minus'] = False    # 正常显示负号
norm[norm <= threshold].plot(style = 'go')    # 正常点
discrete_points = norm[norm > threshold]      # 离群点
discrete_points.plot(style = 'ro')
   
# 标记离群点
for i in range(len(discrete_points)):
    id = discrete_points.index[i]
    n = discrete_points.iloc[i]
    plt.annotate('(%s, %0.2f)' % (id, n), xy = (id, n), xytext = (id, n))

plt.xlabel(u'编号')
plt.ylabel(u'相对距离')
plt.show()
```
